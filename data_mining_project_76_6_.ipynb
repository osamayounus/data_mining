```python
import numpy as np 
import pandas as pd 
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn import metrics
import xgboost as xgb
```

# Read train and test data


```python
trainData=pd.read_csv('train.csv')

o = trainData[trainData['stroke'] == 1]
for i in range(1,20):
    trainData = pd.concat([trainData, o], axis = 0)
```

```python
testData=pd.read_csv('test.csv')
```

# Fill **nulls** in *train* and *test* data

```python
value_r = trainData['bmi'].mean()
trainData['bmi'].fillna(value_r, inplace = True)

value_t = testData['bmi'].mean()
testData['bmi'].fillna(value_t, inplace = True)
```

# drop the attributes we dont need

```python
trainData.drop('id',axis=1,inplace=True)
trainData.drop('work_type' , axis =1 , inplace = True)
trainData.drop('Residence_type' , axis =1 , inplace = True)
```

```python
tid = testData['id']
testData.drop('id',axis=1,inplace=True)
testData.drop('work_type' , axis =1 , inplace = True)
testData.drop('Residence_type' , axis =1 , inplace = True)
```

# **preprocess for train data**

```python
df_encoded = pd.get_dummies(trainData["smoking_status"])
```

```python
trainData=pd.concat([trainData,df_encoded],axis=1)
trainData.drop("smoking_status", axis=1,inplace=True)
```

```python
trainData['gender'].replace({'Male':1,'Female':2,}, inplace=True)
trainData['ever_married'].replace({'Yes':1,'No':0}, inplace=True)
```

```python
trainData['gender']=pd.to_numeric(trainData['gender'],errors='coerce')
trainData.dropna(inplace=True)
```

```python
trainData
```

# **preprocess for test data**

```python
df_encoded = pd.get_dummies(testData["smoking_status"])
```

```python
testData=pd.concat([testData,df_encoded],axis=1)
testData.drop("smoking_status", axis=1,inplace=True)
```

```python
testData['gender'].replace({'Male':1,'Female':2,}, inplace=True)
testData['ever_married'].replace({'Yes':1,'No':0}, inplace=True)
```

```python
testData
```

# **train the model in trainData**

```python
x = trainData.drop("stroke", axis=1)
y = trainData["stroke"]
```

```python
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, stratify=y,random_state=42)
```

```python
smote = SMOTE(random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(x_train, y_train)
```

```python
CLF =RandomForestClassifier(bootstrap = False,
 max_depth = 4,
 min_samples_leaf = 2,
 min_samples_split = 2,
 n_estimators = 950)
```

```python
CLF.fit(X_train_resampled, y_train_resampled)
y_predict = CLF.predict(x_test)
```

```python
metrics.accuracy_score(y_predict, y_test)
```

# **predict the class of test data**

```python
y_predict2=CLF.predict(testData)
```

```python
testData=pd.DataFrame({'id':tid,'stroke':y_predict2})
```
